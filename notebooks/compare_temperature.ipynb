```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_reanalysis = np.loadtxt('../Data/Temperature_reanalysis_2146.txt')
data_measured = pd.read_csv('../Data/Temperature_measured_2146.csv')
```

```python
data_measured['Data-Ora']= pd.to_datetime(data_measured['Data-Ora'])
measured = data_measured[data_measured['Data-Ora'].dt.year == 2010].values[:,2]
```

```python
with plt.xkcd():
    plt.plot(measured+273.15, label = 'Measured')
    plt.plot(data_reanalysis+273.15, label ='Reanalysis')
    plt.ylim(260,310)
    plt.legend()
    plt.show()
```

```python
with plt.xkcd():
    plt.plot(measured-data_reanalysis[:8737], label = 'Difference')
    plt.ylim(-10,+10)
    plt.legend()
    plt.show()

print(np.median(abs(measured-data_reanalysis[:8737])))
```
